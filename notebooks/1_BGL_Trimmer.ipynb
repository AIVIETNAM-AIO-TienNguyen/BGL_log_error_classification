```python
input_path = '/content/BGL.log'
```

```python
total_lines = 0
with open(input_path, 'r', encoding='utf-8', errors='replace') as f:
    for _ in f:
        total_lines += 1
```

```python
file_mb = os.path.getsize(input_path) / 1024**2
print(f"\n📄 File     : {input_path}")
print(f"📦 Size     : {file_mb:.1f} MB")
print(f"📝 Tổng dòng: {total_lines:,}")
```

## 2. Chọn cách cắt

```python
N = 1_500_000

output_path = f"BGL_{N//1000}K_head.log"

with open(input_path, 'r', encoding='utf-8', errors='replace') as fin, \
     open(output_path, 'w', encoding='utf-8') as fout:
    for i, line in enumerate(fin):
        if i >= N: break
        fout.write(line)

out_mb = os.path.getsize(output_path) / 1024**2
print(f"✅ Cách A xong!")
print(f"   Input : {total_lines:,} dòng ({file_mb:.1f} MB)")
print(f"   Output: {N:,} dòng ({out_mb:.1f} MB) → {output_path}")
```

## 3. Kiểm tra file output

```python
# Xem 5 dòng đầu của file vừa cắt
print(f"📋 Preview: {output_path}\n")
with open(output_path, 'r', encoding='utf-8') as f:
    for i, line in enumerate(f):
        if i >= 5: break
        print(f"  {i+1}: {line.rstrip()}")
```

## 4. Download file đã cắt

```python
from google.colab import files
files.download(output_path)
print(f"✅ Đã download: {output_path}")
```
